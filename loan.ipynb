```python
import pandas as pd
import keras
import tensorflow as tf
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
%matplotlib inline
#loading the dataset
df=pd.read_csv('loan_data.csv')
```

```python
#finding missing values in the dataset
df.isnull()
```

```python
df.isnull().sum()
```

```python
#data types
df.dtypes
```

```python

df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)

df['Married'].fillna(df['Married'].mode()[0], inplace=True)

df['Dependents'].fillna(df['Dependents'].mode()[0], inplace=True)

df['Self_Employed'].fillna(df['Self_Employed'].mode()[0], inplace=True)

df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0], inplace=True)

df['Credit_History'].fillna(df['Credit_History'].mode()[0], inplace=True)


```

```python
df['LoanAmount'].fillna(df['LoanAmount'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
#converting the categories into numbers
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
df['Married'] = df['Married'].map({'No': 0, 'Yes': 1})
df['Dependents'] = df['Dependents'].map({'0': 0, '1': 1, '2': 2, '3+': 3})
df['Education'] = df['Education'].map({'Graduate': 1, 'Not Graduate': 0})
df['Self_Employed'] = df['Self_Employed'].map({'No': 0, 'Yes': 1})
df['Property_Area'] = df['Property_Area'].map({'Rural': 0, 'Semiurban': 1, 'Urban': 2})
df['Loan_Status'] = df['Loan_Status'].map({'N': 0, 'Y': 1})
```

```python
#we bring all the variable value in range 0 to 1


for i in df.columns[1:]:
    df[i] = (df[i] - df[i].min()) / (df[i].max() - df[i].min())
```

```python
df.isnull().sum()
```

```python
df = df.drop('Loan_ID', axis=1)
df.shape
```

```python


# Independent variables (features)
X = df.iloc[:, :-1]  # Selecting all columns except the last one

# Dependent variable (target)
y = df.iloc[:, -1]   # Selecting the last column

print(X.shape)
print(y.shape)
```

```python

X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=df['Loan_Status'],random_state=10,test_size=0.2)
```

```python
#import sequential model
from keras.models import Sequential
#importing different layers
from keras.layers import InputLayer, Dense
```

```python
X_train.shape
#number of input feature
input_neurons= X_train.shape[1]
#output will be either yes or no that is either 1 or 0
output_neuron= 1

```

```python
#number of hidden layers
num_of_hidden_layers=2
neuron_hidden_layer1= 15
neuron_hidden_layer2= 10


```

```python
#activation function
model=Sequential()
model.add(InputLayer(input_shape=(input_neurons)))
model.add(Dense(units=neuron_hidden_layer1, activation='tanh'))
model.add(Dense(units=neuron_hidden_layer2, activation='tanh'))
model.add(Dense(units=output_neuron, activation='sigmoid'))
```

```python
model.summary()
```

```python
#number of parameters
input_neurons*neuron_hidden_layer1
```

```python
#adding bias
input_neurons*neuron_hidden_layer1+10
```

```python
#for hidden layer 2
neuron_hidden_layer1*neuron_hidden_layer2
```

```python
#adding bias
neuron_hidden_layer1*neuron_hidden_layer2 + 10
```

```python
neuron_hidden_layer2*output_neuron + 1
```

```python
#compiling the model

model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python

model_history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50)
```

```python
#predictions
predicted_status = model.predict(X_test)
```

```python
#accuracy
accuracyscore=(y_test, predicted_status)
print(accuracyscore)
```
